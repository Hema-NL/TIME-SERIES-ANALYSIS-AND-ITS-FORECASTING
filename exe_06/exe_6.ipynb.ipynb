```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample population dataset (Year vs Population in millions)
data = {
    'Year': np.arange(2000, 2021),
    'Population': [1000, 1020, 1045, 1070, 1090, 1110, 1140, 1170, 1200, 1235,
                   1260, 1290, 1320, 1350, 1385, 1410, 1440, 1470, 1500, 1530, 1560]
}

df = pd.DataFrame(data)

# Moving Average Smoothing
window_size = 3  # You can change this for more/less smoothing
df['Smoothed_Population'] = df['Population'].rolling(window=window_size).mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Population'], label='Original Population', marker='o')
plt.plot(df['Year'], df['Smoothed_Population'], label=f'{window_size}-Year Moving Average', linestyle='--', color='red')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.title('Population Growth with Moving Average Smoothing')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
